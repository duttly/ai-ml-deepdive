```python
from micrograd.engine import Value

a = Value(2.0, label='a')
b = Value(-3.0, label='b')
c = Value(10.0, label='c')

e = a * b; e.label = 'e'
d = e + c; d.label = 'd'

print(f"d.data = {d.data}") 
# Look at how it remembers its history:
print(f"d was created by: {d.op}")
print(d)

d.grad=1.0
d._backward()
e._backward()

print(f"a.grad = {a.grad}")
print(f"b.grad = {b.grad}")
print(f"c.grad = {c.grad}")

# A simple neuron activation
x1 = Value(2.0, label='x1')
w1 = Value(-3.0, label='w1')
x1w1 = x1 * w1; x1w1.label = 'x1*w1'

# Pass through tanh
o = x1w1.tanh(); o.label = 'o'
o.backward()

print(f"o.data = {o.data}")
print(f"x1.grad = {x1.grad}")
print(f"w1.grad = {w1.grad}")
print(o)
```
