```python
from micrograd.engine import Value

a = Value(2.0, label='a')
b = Value(-3.0, label='b')
c = Value(10.0, label='c')

e = a * b; e.label = 'e'
d = e + c; d.label = 'd'

print(f"d.data = {d.data}") 
# Look at how it remembers its history:
print(f"d was created by: {d.op}")
print(d)

d.backward()

print(f"a.grad = {a.grad}")
print(f"b.grad = {b.grad}")
print(f"c.grad = {c.grad}")
```
